## Cleansing Data

**1. Membersihkan Missing Value**

```python
data_raw.isnull().sum()
```

```python
data_raw['Org_Airport'] = data_raw.Org_Airport.fillna("unknown")
```

```python
data_raw['Dest_Airport'] = data_raw.Dest_Airport.fillna("unknown")
```

```python
data_raw.isnull().sum()
```

**2. Menghapus Data Duplikat**

```python
data = data_raw.drop_duplicates()
```

```python
data.dropna()
```

**3. Mengubah Tipe Data**

```python
data['Date'] = pd.to_datetime(data['Date'], dayfirst=True, errors='coerce')

# Cek apakah ada tanggal yang gagal di-parse
if data['Date'].isna().any():
    print("Beberapa tanggal tidak valid:")
    print(data[data['Date'].isna()])
```

```python
print(data['Date'].unique())
```

```python
# Fungsi untuk mengubah integer menjadi format hh:mm
def convert_to_time(value):
    # Pastikan panjang string adalah 4 digit
    str_value = str(value).zfill(4)
    # Ambil bagian jam dan menit
    hours = str_value[:2]
    minutes = str_value[2:]
    # Gabungkan dalam format hh:mm
    return f"{hours}:{minutes}"

# Terapkan fungsi ke kolom time
data['DepTime'] = data['DepTime'].apply(convert_to_time)
data['ArrTime'] = data['ArrTime'].apply(convert_to_time)
data['CRSArrTime'] = data['CRSArrTime'].apply(convert_to_time)

data
```

```python
data.info()
```

**4. Menghapus kolom yang tidak dibutuhkan**

```python
kolom_dihapus = ['Cancelled','CancellationCode','Diverted']
df = data.drop(columns=kolom_dihapus)
df
```

```python
df.shape
```

## Analisis Data

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df.head(5)
```

**1. Analisis Kinerja Maskapai**

```python
carrier_performance = data.groupby('Airline')['ArrDelay'].mean().sort_values(ascending=False)
plt.figure(figsize=(11, 6))
ax = sns.barplot(y=carrier_performance.index, x=carrier_performance.values, palette='viridis')
for i, v in enumerate(carrier_performance.values):
    ax.text(v + 1, i, f"{v:.1f}", color='black', va='center')
plt.title('Rata-rata Keterlambatan Berdasarkan Maskapai')
plt.ylabel('Maskapai')
plt.xlabel('Rata-rata Keterlambatan (menit)')
plt.show()
```

Hasil paling atas menunjukkan maskapai dengan rata-rata keterlambatan tertinggi, sehingga merupakan maskapai dengan performa terburuk. Sebaliknya, hasil paling bawah menunjukkan maskapai dengan rata-rata keterlambatan terendah, yang cenderung lebih tepat waktu.
Dari analisis tersebut, dapat diidentifikasi maskapai yang cenderung menyebabkan keterlambatan penerbangan.

**2. Analisis Hari Terburuk untuk Penerbangan**

```python
days_map = {1: 'Senin', 2: 'Selasa', 3: 'Rabu', 4: 'Kamis', 5: 'Jumat', 6: 'Sabtu', 7: 'Minggu'}
data['DayName'] = data['DayOfWeek'].map(days_map)
day_performance = data.groupby('DayName')['ArrDelay'].mean().reindex(days_map.values())
plt.figure(figsize=(8, 5))
ax = sns.barplot(x=day_performance.index, y=day_performance.values, palette='coolwarm')
for i, v in enumerate(day_performance.values):
    ax.text(i, v + 0.5, f"{v:.1f}", color='black', ha='center')
plt.title('Rata-rata Keterlambatan Berdasarkan Hari dalam Seminggu')
plt.xlabel('Hari dalam Seminggu')
plt.ylabel('Rata-rata Keterlambatan (menit)')
plt.xticks(rotation=45)

plt.show()
```

analisis ini membandingkan rata-rata keterlambatan per hari dalam seminggu secara keseluruhan dalam dataset, bukan membandingkan antar-minggu. Analisis hari dengan rata-rata keterlambatan tertinggi menunjukkan waktu dalam seminggu yang cenderung lebih sibuk atau mengalami gangguan operasional.

**3. Analisis Keterlambatan Berdasarkan Rute Penerbangan**

```python
if 'Route' not in df.columns:
    df['Route'] = df['Origin'] + ' â†’ ' + df['Dest']
route_performance = df.groupby('Route')['ArrDelay'].mean().sort_values(ascending=False).head(10)
plt.figure(figsize=(12, 6))
ax = sns.barplot(x=route_performance.values, y=route_performance.index, palette='plasma')
for i, v in enumerate(route_performance.values):
    ax.text(v + 1, i, f"{v:.1f}", color='black', va='center')
plt.title('Rute Penerbangan dengan Rata-rata Keterlambatan Tertinggi')
plt.xlabel('Rata-rata Keterlambatan (menit)')
plt.ylabel('Rute Penerbangan')
plt.show()
```

Hasil menunjukkan rata-rata rute penerbangan yang mengalami keterlambatan. Seperti hasil yang tertinggi rute EGE â†’ MIA memiliki rata -rata keterlambatan sebesar 320 menit, yang berarti rute tersebut sering mengalami keterlambatan signifikan dibandingkan rute lain. Namun, bukan berarti setiap penerbangan pada rute tersebut selalu terlambat.

**4. Analisis Faktor Penyebab Keterlambatan**

```python
causes = ['CarrierDelay', 'WeatherDelay', 'NASDelay', 'SecurityDelay', 'LateAircraftDelay']
delay_causes = df[causes].sum()
plt.figure(figsize=(8, 8))
delay_causes.plot(kind='pie', autopct='%1.1f%%', startangle=140)
plt.title('Faktor Penyebab Keterlambatan')
plt.ylabel('')
plt.show()
```

penjelasan penyebab keterlambatan:

CarrierDelay: Disebabkan oleh masalah internal maskapai seperti kru, perawatan pesawat, atau pengisian bahan bakar.

WeatherDelay: Disebabkan oleh kondisi cuaca buruk seperti badai, angin kencang, atau kabut.

NASDelay: Masalah pada National Airspace System (NAS), seperti kendala kontrol lalu lintas udara atau infrastruktur bandara.

SecurityDelay: Penundaan akibat masalah keamanan seperti pemeriksaan tambahan.

LateAircraftDelay: Pesawat yang datang terlambat dari penerbangan sebelumnya.

**5. Analisis Bandara Asal dan Tujuan Terburuk**

```python
# Group and sort the data as in your original code
origin_performance = df.groupby('Org_Airport')['ArrDelay'].mean().sort_values(ascending=False).head(10)
dest_performance = df.groupby('Dest_Airport')['ArrDelay'].mean().sort_values(ascending=False).head(10)

# Create the figure and subplots
plt.figure(figsize=(18, 5))

# First subplot for origin airport performance (horizontal bar chart)
ax1 = plt.subplot(1, 2, 1)
ax1 = sns.barplot(x=origin_performance.values, y=origin_performance.index, palette='rocket')
for i, v in enumerate(origin_performance.values):
    ax1.text(v + 1, i, f"{v:.1f}", color='black', va='center')
plt.title('Bandara Asal dengan Rata-rata Keterlambatan Tertinggi')
plt.xlabel('Rata-rata Keterlambatan (menit)')
plt.ylabel('Bandara Asal')

plt.tight_layout()
plt.show()
```

bandara dengan hasil paling atas menunjukkan rata-rata keterlambatan untuk semua penerbangan yang berangkat dari bandara tersebut. Artinya, bandara tersebut memiliki kecenderungan menjadi sumber keterlambatan.

```python
plt.figure(figsize=(18, 5))

# Second subplot for destination airport performance (horizontal bar chart)
ax2 = plt.subplot(1, 2, 2)
ax2 = sns.barplot(x=dest_performance.values, y=dest_performance.index, palette='mako')
for i, v in enumerate(dest_performance.values):
    ax2.text(v + 1, i, f"{v:.1f}", color='black', va='center')
plt.title('Bandara Tujuan dengan Rata-rata Keterlambatan Tertinggi')
plt.xlabel('Rata-rata Keterlambatan (menit)')
plt.ylabel('Bandara Tujuan')

plt.tight_layout()
plt.show()
```

Bandara dengan hasil paling atas menunjukkan rata-rata keterlambatan untuk semua penerbangan yang tiba di bandara tersebut. Artinya, bandara tersebut memiliki kecenderungan mengalami keterlambatan saat menerima kedatangan pesawat, yang dapat disebabkan oleh faktor internal bandara, seperti kondisi operasional, cuaca, atau kepadatan lalu lintas udara.

**6. Tren Keterlambatan Berdasarkan Bulan**

```python
df['Date'] = pd.to_datetime(df['Date'])
date_performance = df.groupby(df['Date'].dt.to_period('M'))['ArrDelay'].mean()
plt.figure(figsize=(12, 6))
date_performance.plot(kind='line', marker='o', color='b')
plt.title('Tren Rata-rata Keterlambatan Berdasarkan Bulan')
plt.xlabel('Bulan')
plt.ylabel('Rata-rata Keterlambatan (menit)')
plt.grid()
plt.show()
```

Hasil menunjukkan bahwa rata-rata keterlambatan pada bulan Februari adalah yang tertinggi, tetapi ini tidak berarti semua penerbangan pada bulan tersebut mengalami penundaan. Mungkin ada beberapa kejadian khusus, seperti cuaca buruk atau masalah lain, yang meningkatkan rata-rata keterlambatan pada bulan tersebut.
